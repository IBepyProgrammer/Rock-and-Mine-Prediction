```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("sonar_csv.csv", header = None)
```

### We then group the data by the rocks and mines columns

```python
df.groupby(60).mean()
```

```python
X = df.drop(columns = 60, axis=1)
y = df[60]
```

```python
print(X)
```

```python
print(y)
```

```python
# stratify = y gices equal number of values(i.e Rocks & Mines) to training anf test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, stratify = y, random_state = 1)
```

#### We then trtain our model using logistic regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

### we then find the accuracy score on the train data

```python
X_train_pred = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_pred, y_train)

print("Accuracy on training data is: ", train_data_accuracy *100,"%")
```

### We then find the accuracy score on the test data

```python
X_test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred, y_test)

print("Accuracy on training data is: ", test_data_accuracy *100,"%")
```
